```python
import math
import numpy as np
import matplotlib.pyplot as plt
from ideal_functions import BlimpLift
```

```python
# Simulation Parameters
payload = 135  # grams
helium_volume = 141  # liters
bladder_fill = 10  # liters
temperature_range = np.linspace(-5, 40, 500)  # Celsius
altitude_range = np.linspace(0, 500, 1000)  # meters

blimp = BlimpLift(payload, helium_volume, bladder_fill)

# Calculate lift for different temperatures and altitudes
lift_results = []
for temperature in temperature_range:
    temp_lift = []
    for altitude in altitude_range:
        net_lift, _ = blimp.calculate_lift(temperature, altitude)
        temp_lift.append(net_lift)
    lift_results.append(temp_lift)

# Convert results to a NumPy array for plotting
lift_results = np.array(lift_results)

# Visualization
plt.figure(figsize=(10, 6))
temperature_idx = [0, 10, 20, 30, 40, 49]  # Indices of specific temperatures to plot
for idx in temperature_idx:
    plt.plot(altitude_range, lift_results[idx], label=f'Temp = {temperature_range[idx]:.1f}C')

plt.title('Net Lift Force vs Altitude at Different Temperatures')
plt.xlabel('Altitude (m)')
plt.ylabel('Net Lift Force (N)')
plt.legend()
plt.grid(True)
plt.show()
```
